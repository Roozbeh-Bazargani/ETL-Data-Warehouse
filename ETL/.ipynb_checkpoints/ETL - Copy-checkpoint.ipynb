```python
import psycopg2
import psycopg2.extras
from optparse import OptionParser, OptionGroup
import sys
```

```python
# Connect to an existing database
conn1 = psycopg2.connect("dbname=DB-project user=postgres password=admin")
conn2 = psycopg2.connect("dbname=Data-Warehouse user=postgres password=admin")
```

```python
cur1 = conn1.cursor()
cur2 = conn2.cursor()

conn1.autocommit = True
conn2.autocommit = True
```

# Pipeline

```python
def columns(cursor, table):
    cursor.execute("""
    SELECT COLUMN_NAME
    FROM INFORMATION_SCHEMA.COLUMNS
    WHERE TABLE_NAME = '{table}'
    ORDER BY ORDINAL_POSITION
    """.format(table=table))
    
def attributes(table):
    with conn2.cursor() as cursor2:
        columns(cursor2, table)
        cols = cursor2.fetchall()
        attributes = []
        for col in cols:
            attributes.append(col[0])
        return attributes
```

```python
def trigger(table):
    atts = attributes(table)
    s = ''
    s_old = ''
    for att in atts:
        s += att + ', '
    for att in atts:
        s_old += 'OLD.' + att + ', '
    query = """CREATE FUNCTION {table}_history() RETURNS trigger AS ${table}_history$
    BEGIN
        IF (TG_OP = 'UPDATE') THEN
            INSERT INTO {table}_old(""".format(table=table)
    query += s
    query += """change_date, change)
            VALUES("""
    query += s_old
    query += """CURRENT_DATE, 'update');
        ELSEIF (TG_OP = 'DELETE') THEN
            INSERT INTO {table}_old(""".format(table=table)
    query += s
    query += """change_date, change)
            VALUES("""
    query += s_old
    query += """CURRENT_DATE, 'delete');
        END IF;
        RETURN NULL; -- result is ignored since this is an AFTER trigger
    END;
    ${table}_history$ LANGUAGE plpgsql;

    CREATE TRIGGER {table}_history AFTER UPDATE OR DELETE ON {table}
    FOR EACH ROW EXECUTE FUNCTION {table}_history();
    """.format(table=table)
    
    return query
```

```python
print(trigger('person'))
```

```python
def tables(cursor):
    cursor.execute("SELECT tablename FROM pg_tables WHERE schemaname='public'")
    tables = []
    for row in cursor.fetchall():
        tables.append(row[0])
    return tables

with conn1.cursor() as cursor1:
    tables = tables(cursor1)
    print('list of tables:\n', tables)
    for table in tables:
        print('\n\n', trigger(table), '\n\n')
```
